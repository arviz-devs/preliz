```python
%matplotlib widget
import pytest
import ipytest
ipytest.autoconfig()

import numpy as np
from preliz import ppa, HalfNormal, Normal
```

```python
%%ipytest

@pytest.fixture
def fmodel():
    def a_preliz_model():
        a = Normal(0, 10).rvs()
        b = HalfNormal(10).rvs()
        y = Normal(a, b).rvs(100)
        return a, b, y
    return a_preliz_model

@pytest.mark.parametrize("draws, references, boundaries, target, engine", [
    (1000, 0, (-np.inf, np.inf), None, "preliz"), # target set to  default
    (2000, (0, 100), (0, np.inf), (500, 300), "preliz"), # target tuple with mean=500 and std_deviation=300
    (4000, 0.5, (-np.inf, 0), (400, 600), "preliz"), # target tuple with mean=400 and std_deviation=600
])
def test_ppa(fmodel, draws, references, boundaries, target, engine):
    ppa(fmodel, draws, references, boundaries, target, engine)
```
